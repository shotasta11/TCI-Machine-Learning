```python
import pandas as pd
import numpy as np
import os 
```

```python
df = pd.read_excel('ERA5_Gianyar.xlsx')

df
```

```python
df.describe()
```

```python
#rata-rata tiap parameter dalam rentang waktu 1 hari
avg_perhari = df.groupby(np.arange(len(df))//24).mean()

avg_perhari
```

```python
#Membuat Data frame baru dengan memasukan parameter TCI
df_TCI = avg_perhari.assign(TCI = avg_perhari.t2m[:60]*4 + avg_perhari.d2m[:60] + avg_perhari.windspeed_km_h[:60]+ avg_perhari.tp_mm[:60]*2)
df_TCI
```

```python
df_TCI.drop(df_TCI.columns[0:7], axis=1, inplace=True)
df_TCI
```

```python
df_TCI.to_csv("Gianyar_tci.csv")
```
